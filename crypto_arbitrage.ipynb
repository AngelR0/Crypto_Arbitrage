```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
#Reading in the CSV bitstamp.csv
#The index column is set to "Timestamp"
#Set the parse_dates and infer_datetime_format parameters to True

bitstamp = pd.read_csv(Path("Resources/bitstamp.csv"),
                      index_col="Timestamp",
                      parse_dates=True,
                      infer_datetime_format=True)
```

```python
print("Bitstamp \n")

#printing the first rows
print(bitstamp.head(), "\n")

#printing the last rows
print(bitstamp.tail())
```

```python
#Reading in the CSV coinbase.csv
#The index column is set to "Timestamp"
#Set the parse_dates and infer_datetime_format parameters to True

coinbase = pd.read_csv(Path("Resources/coinbase.csv"),
                      index_col="Timestamp",
                      parse_dates=True,
                      infer_datetime_format=True)
```

```python
print("Coinbase\n")

#printing the first rows
print(coinbase.head(), "\n")

#printing the last rows
print(coinbase.tail(), "\n")
```

```python
#Checking for NaN values
print("Bitstamp\n",bitstamp.isnull().sum(), "\n")

print("Coinbase\n",coinbase.isnull().sum())
```

```python
# Dropping the NaN for both
bitstamp.dropna(inplace=True)

coinbase.dropna(inplace=True)
```

```python
#Checking if the NaN values dropped
print("Bitstamp\n")
print(bitstamp.isnull().sum(), "\n")

print("Coinbase\n")
print(coinbase.isnull().sum())
```

```python
# Removing the dollar sign '$', from the column "Close"

bitstamp.loc[:, "Close"] = bitstamp.loc[:, "Close"].str.replace('$', '')

coinbase.loc[:, "Close"] = coinbase.loc[:, "Close"].str.replace('$', '')
```

```python
bitstamp.loc[:, "Close"] = bitstamp.loc[:, "Close"].astype(float)

coinbase.loc[:, "Close"] = coinbase.loc[:, "Close"].astype(float)
```

```python
# Checking for duplicates
print("Bitstamp duplicates: ", bitstamp.duplicated().sum(), "\n")

print("Coinbase duplicates: ", coinbase.duplicated().sum(), )
```

```python
#Slicing just the "Close" column for both Bitstamp and Coinbase

bitstamp_sliced = bitstamp.loc[:, "Close"]

coinbase_sliced = coinbase.loc[:, "Close"]
```

```python
print("Bitstamp 'Closed': \n", bitstamp_sliced.head(), "\n")

print("Coinbase 'Closed': \n", coinbase_sliced, "\n")
```

```python
bitstamp_sliced.describe()
```

```python
coinbase_sliced.describe()
```

```python
bitstamp_sliced.plot(figsize=(15,10), title=("Bitstamp Closing Prices"), color="blue")
```

```python
coinbase_sliced.plot(figsize=(15,10), title=("Coinbase Closing Prices"), color="orange")
```

```python
bitstamp_sliced.plot(legend=True, figsize=(15,15), title=("Bitstamp & Coinbase Closing Prices"), color="blue", label="Bitstamp")
coinbase_sliced.plot(legend=True, figsize=(15,15), color="orange", label="Coinbase")
```

```python
bitstamp_sliced["2018-01-01":"2018-02-01"].plot(legend=True, figsize=(15,15), title="Early Time Period Bitstamp & Coinbase Closing Prices",label="Bitstamp", color="blue")

coinbase_sliced["2018-01-01":"2018-02-01"].plot(legend=True, figsize=(15,15), label="Coinbase", color="orange")
```

```python
bitstamp_sliced["2018-03-01":"2018-04-01"].plot(legend=True, figsize=(15,15), title="Later Time Period Bitstamp & Coinbase Closing Prices",label="Bitstamp", color="blue")

coinbase_sliced["2018-03-01":"2018-04-01"].plot(legend=True, figsize=(15,15), label="Coinbase", color="orange")
```

From the months of January through the 1st of February the degree of spread change was high but in the later month of March through the 1st of April, the degree of spread changed decreased.

# Arbitrage

```python
#Selecting an early date dataset

bitstamp_sliced["2018-01-16"].plot(legend=True, figsize=(15,15), title="2018-01-16 Bitstamp & Coinbase",label="Bitstamp", color="blue")

coinbase_sliced["2018-01-16"].plot(legend=True, figsize=(15,15), label="Coinbase", color="orange")
```

```python
early_spread = coinbase_sliced.loc["2018-01-16"] - bitstamp_sliced.loc["2018-01-16"]

early_spread.describe()
```

```python
early_spread.plot(kind='box', title="Early Spread Dataset")
```

```python
#Middle date dataset

bitstamp_sliced["2018-02-02"].plot(legend=True, figsize=(15,15), title="2018-02-02 Bitstamp & Coinbase",label="Bitstamp", color="blue")

coinbase_sliced["2018-02-02"].plot(legend=True, figsize=(15,15), label="Coinbase", color="orange")
```

```python
middle_spread = bitstamp_sliced.loc["2018-02-02"] - coinbase_sliced.loc["2018-02-02"]

middle_spread.describe()
```

```python
middle_spread.plot(kind='box', title="Middle Spread Dataset")
```

```python
#Late date dataset

bitstamp_sliced["2018-03-30"].plot(legend=True, figsize=(15,15), title="2018-03-30 Bitstamp & Coinbase",label="Bitstamp", color="blue")

coinbase_sliced["2018-03-30"].plot(legend=True, figsize=(15,15), label="Coinbase", color="orange")
```

```python
late_spread = bitstamp_sliced.loc["2018-03-30"] - coinbase_sliced.loc["2018-03-30"]

late_spread.describe()
```

```python
late_spread.plot(kind='box', title="Late Spread Dataset")
```

```python
#Calculating the early arbitrage spread

early_arbitrage_spread = coinbase_sliced.loc["2018-01-16"] - bitstamp_sliced.loc["2018-01-16"]

early_arbitrage_spread = early_arbitrage_spread.dropna()

early_arbitrage_spread.describe()
```

```python
early_spread_return = early_arbitrage_spread[early_arbitrage_spread>0] / bitstamp_sliced["2018-01-16"]

early_spread_return
```

```python
profitable_early_trades = early_spread_return[early_spread_return>.01]

profitable_early_trades.head()
```

```python
profitable_early_trades.describe()
```

```python
profit_early = profitable_early_trades * bitstamp_sliced["2018-01-16"]

profit_early = profit_early.dropna()

profit_early
```

```python
profit_early.describe()
```

```python
profit_early.plot(title = "Early Profitable Trades")
```

```python
profit_early.sum()
```

```python
cumulative_early_profit = profit_early.cumsum()
```

```python
cumulative_early_profit.plot(title = "Cumulative Sum of Early Date")
```

```python
#Calculating the middle arbitrage spread

middle_arbitrage_spread = bitstamp["Close"].loc["2018-02-02"] - coinbase["Close"].loc["2018-02-02"]

middle_arbitrage_spread.describe()
```

```python
middle_spread_return = middle_arbitrage_spread[middle_arbitrage_spread>0] / bitstamp_sliced.loc["2018-02-02"]

middle_spread_return
```

```python
profitable_middle_trades = middle_spread_return[middle_spread_return>.01]

profitable_middle_trades.head()
```

```python
profitable_middle_trades.describe()
```

```python
profit_middle = profitable_middle_trades * bitstamp_sliced.loc["2018-02-02"]

profit_middle = profit_middle.dropna()

profit_middle
```

```python
profit_middle.describe()
```

```python
profit_middle.plot(title = "Middle Profitable Trades")
```

```python
profit_middle.sum()
```

```python
cumulative_middle_profit = profit_middle.cumsum()

cumulative_middle_profit.plot(title = "Cumulative Sum of Middle Date")
```

```python
#Calculating the late arbitrage spread

late_arbitrage_spread = coinbase["Close"].loc["2018-03-30"] - bitstamp["Close"].loc["2018-03-30"]

late_arbitrage_spread.describe()
```

```python
late_spread_return = late_arbitrage_spread[late_arbitrage_spread>0] / bitstamp_sliced["2018-03-30"]

late_spread_return = late_spread_return.dropna()

late_spread_return.head()
```

```python
profitable_late_trades = late_spread_return[late_spread_return>.01]

profitable_late_trades.describe()
```

```python
profit_late = profitable_late_trades * coinbase_sliced.loc["2018-03-30"]

profit_late = profit_late.dropna()

profit_late.describe()
```

```python
profit_late.plot(title = "Late Profitable Trades")
```

```python
profit_late.sum()
```

```python
cumulative_late_profit = profit_late.cumsum()

cumulative_late_profit.plot(title = "Cumulative Sum of Late Date")
```

As it is shown in all cumulative sums of early, middle, and late dates. We are shown that in the beginning we have a higher possibility to have a 1% return. In the middle date, 1% return is still likely but less probable. Finally, in later dates it was unlikely to recieve a 1% return.
