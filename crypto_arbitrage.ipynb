```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
#Reading in the CSV bitstamp.csv
#The index column is set to "Timestamp"
#Set the parse_dates and infer_datetime_format parameters to True

bitstamp = pd.read_csv(Path("Resources/bitstamp.csv"),
                      index_col="Timestamp",
                      parse_dates=True,
                      infer_datetime_format=True)
```

```python
print("Bitstamp \n")

#printing the first rows
print(bitstamp.head(), "\n")

#printing the last rows
print(bitstamp.tail())
```

```python
#Reading in the CSV coinbase.csv
#The index column is set to "Timestamp"
#Set the parse_dates and infer_datetime_format parameters to True

coinbase = pd.read_csv(Path("Resources/coinbase.csv"),
                      index_col="Timestamp",
                      parse_dates=True,
                      infer_datetime_format=True)
```

```python
print("Coinbase\n")

#printing the first rows
print(coinbase.head(), "\n")

#printing the last rows
print(coinbase.tail(), "\n")
```

```python
#Checking for NaN values
print("Bitstamp\n",bitstamp.isnull().sum(), "\n")

print("Coinbase\n",coinbase.isnull().sum())
```

```python
# Dropping the NaN for both
bitstamp.dropna(inplace=True)

coinbase.dropna(inplace=True)
```

```python
#Checking if the NaN values dropped
print("Bitstamp\n")
print(bitstamp.isnull().sum(), "\n")

print("Coinbase\n")
print(coinbase.isnull().sum())
```

```python
# Removing the dollar sign '$', from the column "Close"

bitstamp.loc[:, "Close"] = bitstamp.loc[:, "Close"].str.replace('$', '')

coinbase.loc[:, "Close"] = coinbase.loc[:, "Close"].str.replace('$', '')
```

```python
bitstamp.loc[:, "Close"] = bitstamp.loc[:, "Close"].astype(float)

coinbase.loc[:, "Close"] = coinbase.loc[:, "Close"].astype(float)
```

```python
# Checking for duplicates
print("Bitstamp duplicates: ", bitstamp.duplicated().sum(), "\n")

print("Coinbase duplicates: ", coinbase.duplicated().sum(), )
```

```python
#Slicing just the "Close" column for both Bitstamp and Coinbase

bitstamp_sliced = bitstamp.loc[:, "Close"]

coinbase_sliced = coinbase.loc[:, "Close"]
```

```python
print("Bitstamp 'Closed': \n", bitstamp_sliced.head(), "\n")

print("Coinbase 'Closed': \n", coinbase_sliced, "\n")
```

```python
bitstamp_sliced.describe()
```

```python
coinbase_sliced.describe()
```

```python
bitstamp_sliced.plot(figsize=(15,10), title=("Bitstamp Closing Prices"), color="blue")
```

```python
coinbase_sliced.plot(figsize=(15,10), title=("Coinbase Closing Prices"), color="orange")
```

```python
bitstamp_sliced.plot(legend=True, figsize=(15,15), title=("Bitstamp & Coinbase Closing Prices"), color="blue", label="Bitstamp")
coinbase_sliced.plot(legend=True, figsize=(15,15), color="orange", label="Coinbase")
```

```python
bitstamp_sliced["2018-01-01":"2018-02-01"].plot(legend=True, figsize=(15,15), title="Early Time Period Bitstamp & Coinbase Closing Prices",label="Bitstamp", color="blue")

coinbase_sliced["2018-01-01":"2018-02-01"].plot(legend=True, figsize=(15,15), label="Coinbase", color="orange")
```

```python
bitstamp_sliced["2018-03-01":"2018-04-01"].plot(legend=True, figsize=(15,15), title="Later Time Period Bitstamp & Coinbase Closing Prices",label="Bitstamp", color="blue")

coinbase_sliced["2018-03-01":"2018-04-01"].plot(legend=True, figsize=(15,15), label="Coinbase", color="orange")
```

From the months of January through the 1st of February the degree of spread change was high but in the later month of March through the 1st of April, the degree of spread changed decreased.
